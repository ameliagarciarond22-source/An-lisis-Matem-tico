###Conjunto de Cantor.

#### Contexto Histórico y Origen

A finales del siglo XIX, los matemáticos creían que si un conjunto era "grande" (tenía muchos puntos), debía ocupar "espacio" (tener longitud). Georg Cantor demostró que esto era falso. Creó un objeto que es tan grande como todos los números reales, pero que ocupa exactamente cero espacio.

#### La Paradoja de la Medida vs. el Cardinal

Este es el "corazón" del análisis en este tema:

* Medida de Lebesgue: La construcción consiste en remover progresivamente los tercios centrales. La suma de las longitudes de los intervalos removidos es una serie geométrica:
$$\sum_{n=1}^{\infty} \frac{2^{n-1}}{3^n} = \frac{1}{3} \sum_{n=0}^{\infty} \left(\frac{2}{3}\right)^n = \frac{1}{3} \left( \frac{1}{1 - 2/3} \right) = 1$$ Como el intervalo original $[0,1]$ tiene longitud 1 y le quitamos una longitud total de 1, lo que queda tiene medida cero.

Si sumas todos los trozos que quitas, la suma da 1. Es decir, "quitas todo el intervalo original".

* Puntos (Cardinalidad): Aunque "vaciaste" el intervalo, los puntos que quedan (como los extremos de los segmentos) son tantos que no se pueden contar. Sigue siendo un conjunto no numerable.

#### Propiedades Topológicas

En análisis,se le llama polvo de Cantor porque es un conjunto "agujereado" al extremo, se describe como un conjunto "patológico" por poseer al mismo tiempo estas características:

* Cerrado y Acotado: Por el teorema de Heine-Borel, es un conjunto compacto.

* Denso en ninguna parte: Su interior es vacío, lo que significa que no contiene ningún intervalo abierto.
No tiene "cuerpo": No contiene ningún intervalo abierto, por pequeño que sea. Si te paras en un punto del conjunto y caminas un milímetro, ya te saliste de él.

* Perfecto: Es un conjunto cerrado donde cada uno de sus puntos es un punto límite.
A pesar de ser puro polvo, no tiene puntos aislados. Cada punto del conjunto está rodeado de infinitos otros puntos del mismo conjunto a una distancia infinitesimal.

* Totalmente desconectado: Las únicas componentes conexas son los puntos individuales.

```python
import matplotlib.pyplot as plt

class Cantor:
  def Grafica(self, pi, pf, nivel):
    xd = [pi, pf]
    yd = [1/ (nivel), 1/(nivel)]
    plt.plot(xd,yd,linewidth=2,color='purple')

    puntos = []
    incre = (1/3)** nivel
    puntos.append([pi, pi+incre])
    puntos.append([pf-incre, pf])
    return puntos

  def Construye(self, nivel):
    puntos = []
    puntos.append([0,1])
    for i in range(nivel):
      Npuntos = []
      for pdz in puntos:
        Npuntos.extend(self.Grafica(pdz[0], pdz[1], i+1))
      puntos = Npuntos

    plt.show()

cant = Cantor()
cant.Construye(5)
```
