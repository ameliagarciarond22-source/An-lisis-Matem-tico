###Conjunto de Cantor.

#### Contexto Histórico y Origen
El conjunto fue introducido por Georg Cantor en su artículo de 1883, "Über unendliche, lineare Punktmannichfaltigkeiten". Surgió en el marco de sus investigaciones sobre series trigonométricas y conjuntos de puntos excepcionales, desafiando la noción intuitiva de que la "cantidad" de puntos (cardinalidad) y la "extensión" (medida) debían estar ligadas proporcionalmente.

#### Paradoja de la Medida vs. Cardinalidad

Ilustra la diferencia entre la Medida de Lebesgue y la Cardinalidad.


* Medida de Lebesgue ($m(C) = 0$):
La construcción consiste en remover progresivamente los tercios centrales. La suma de las longitudes de los intervalos removidos es una serie geométrica:
$$\sum_{n=1}^{\infty} \frac{2^{n-1}}{3^n} = \frac{1}{3} \sum_{n=0}^{\infty} \left(\frac{2}{3}\right)^n = \frac{1}{3} \left( \frac{1}{1 - 2/3} \right) = 1$$ Como el intervalo original $[0,1]$ tiene longitud 1 y le quitamos una longitud total de 1, lo que queda tiene medida cero.

* Cardinalidad ($|C| = |\mathbb{R}|$):
A pesar de tener medida nula, el conjunto es no numerable.

#### Propiedades Topológicas

En el análisis de espacios métricos, el conjunto de Cantor se describe como un conjunto "patológico" por poseer al mismo tiempo estas características:

* Cerrado y Acotado: Por el teorema de Heine-Borel, es un conjunto compacto.
* Perfecto: Es un conjunto cerrado donde cada uno de sus puntos es un punto límite (no tiene puntos aislados).
* Denso en ninguna parte: Su interior es vacío, lo que significa que no contiene ningún intervalo abierto.
* Totalmente desconectado: Las únicas componentes conexas son los puntos individuales.

```python
import matplotlib.pyplot as plt

def GCantor(pi, pf, nivel):
  #pi= punto inicial del intervalo, pf= punto final, nivel = número de iteración
  puntos = []
  incre = (1/3)** nivel
  xs = [pi, pi+incre]
  ys = [1/(nivel+1),1/(nivel+1)]
  plt.plot(xs, ys, linewidth=1, color='purple')
  puntos.append(xs)

  xs = [pf-incre, pf]
  ys = [1/(nivel+1),1/(nivel+1)]
  plt.plot(xs, ys, linewidth=1, color='purple')
  puntos.append(xs)

  return puntos

puntos = [[0, 1]]
puntos = GCantor(0, 1, 0)
for i in range(5) :
  Npuntos = []
  for pdz in puntos:
    Npuntos.extend(GCantor(pdz[0], pdz[1], i))
  puntos = Npuntos
plt.show()
```
